```python
#1차원 배열 형태인 시리즈(Series), 2차원 배열 형태인 데이터프레임(DataFrame)
import pandas as pd
import os

df = pd.DataFrame([['kim',27,92],
                   ['lee',33,98],
                   ['park',19,87]])
df2 = pd.DataFrame({'name':['kim','lee','park'],
                     'age':[27,33,19],
                     'score':[92,98,87]})
print(df)
print(df2)

currentpath = os.getcwd()
print(currentpath)

df = pd.read_csv('data.csv',encoding='UTF-8') #encoding이하는 생략가능 불러온 파일의 글자가 깨졌을 경우에 사용.
print(df)

url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/lol.csv'
df = pd.read_csv(url)
print(df)
print(df.shape)

df = pd.read_excel('data.xlsx')
print(df)

#head(),tail() : 데이터프레임의 상위/하위 데이터 몇 개만 보기,괄호 빈칸일 시 5개씩 보여줌.
print(df.head())

#info() : 데이터프레임에 대한 기본 정보를 출력한다. 10 non-null : 빠진 데이터가 없다/ 7 non-null : 10행 중에 3개의 행에 데이터가 비어있다.
#         결측치(missing value), 비결측치
df.info()

#특정 칼럼의 데이터 불러오기
print(df['이름'],'\n')
print(df.반,'\n')
print(df['이름'][0:3]) #특정 행

#loc, iloc
print(df.loc[2])
print(df.loc[0:3,['이름']],'\n')
print(df.iloc[1,1])
print(df.iloc[0:3,0:2])

#조건 연산
print(df['국어']>=80)
print(df[df['국어']>=80])
print(df.query('국어>=80'))
print(df[(df['국어']>=80) & (df['수학']>=90)]) #and
print(df[(df['국어']>=80) | (df['수학']>=90)]) #or, pipeline

#정렬
print(df.sort_values(by=['국어'],ascending=False)) #ascending=오름차순

print(list(df.index))
print(list(df.columns))
print(df.size) #행x열

print(df.describe())

print(df['성별'].value_counts())

print(df['국어'].sum())
print(df['국어'].mean())

print(df.groupby('반').sum())

#결측치 확인
print(df.isna())
print(df.isna().sum())

결측치 값 채우기: inplace=기존의 데이터를 바꿀지 말지 / True:바꿈, False:안바꿈.

df.dropna() : 데이터 삭제 / (4)은 패스

#새로운 colunms 추가
df['총점'] = df['국어'] + df['수학']
print(df)

df.drip() : 데이터프레임의 행,칼럼 삭제
print(df.drop('확인여부',axis=1)) #axis가 뭐지?

df = pd.read_excel('BoxOffice.xlsx')
print('#1')
print(df.info())
print('#2')
print(df['대표국적'].value_counts(),'\n')
print('#3')
print(df[df['대표국적'] == '영국'],'\n')
print('#4')
print(df.groupby('대표국적')['순위'].min())
```
